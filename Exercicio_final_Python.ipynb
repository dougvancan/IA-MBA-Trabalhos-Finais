```python
#1 - Leia o arquivo lemonades.csv usando pandas

import pandas as pd

lemonade = pd.read_csv('Lemonade2016(2).csv', sep=";")
lemonade.head(10)
```

```python
#3 - Realize a limpeza do dataset:
#- verifique se existe valores nulos

print("Colunas elementos null:\n", lemonade.isnull().any())

print("\nNull:\n", lemonade[lemonade.isnull().any(axis=1)])
```

```python
#- remova duplicatas
print("\nDuplicados\n", lemonade.duplicated())
lemonade.drop_duplicates()
```

```python
#- trate os elementos faltantes (interpolação e media)
#De acordo com análise, vou substituir o null do date por 07/08/2016
#e o Leaflets pela média

lemonade[lemonade["Date"].isnull()] = lemonade[lemonade["Date"].isnull()].fillna("07/08/2016")
lemonade[lemonade["Leaflets"].isnull()] = lemonade[lemonade["Leaflets"].isnull()].fillna(lemonade["Leaflets"].mean())
```

```python
#4 - Adicione uma coluna chamada "Sales" que contém o total de vendas de limão e laranja
from pandas import Series, DataFrame

lemonade_df = DataFrame(lemonade)
lemonade_df["Sales"] = lemonade_df["Lemon"] + lemonade_df["Orange"]
lemonade_df
```

```python
#5 - Adicione uma coluna chamada "Revenue" que calcula o lucro (venda*preço)

lemonade_df["Revenue"] = lemonade_df["Sales"] * lemonade_df["Price"]
lemonade_df
```

```python
#6 - retorne o lucro total
lemonade_df["Revenue"].sum()
```

```python
#7 - escreva uma função que receba dois parâmetros, dataset e temp(int). Se max for 1, ele
#retorna a máxima temperatura observada no conjunto de dados; se 0, retorna a média do
#período observado; se -1, retorna a temperatura mínima

def temperatura(ds, temp):
    if temp == 1:   
        return ds["Temperature"].max()
    elif temp == 0:
        return ds["Temperature"].mean()
    elif temp == -1:
        return ds["Temperature"].min()

```

```python
print("Igual a 1: ", temperatura(lemonade_df, 1))
print("Igual a 0: ", temperatura(lemonade_df, 0))
print("Igual a -1: ", temperatura(lemonade_df, -1))
```

```python
#8 - escreva uma função que receba dois parametros (dataset, localização) e retorne o dataset
#com o preço do limão e laranja ajustados em 15% se a localização for 'Park' ou ajustados em
#10% se a localização for 'Beach'

def preco(ds, local):
    if local == "Park":
        ds["Price"] = ds["Price"] * 1.15
    elif local == "Beach":
        ds["Price"] = ds["Price"] * 1.1
    return ds
        
```

```python
print(preco(lemonade_df, "Park"))
```

```python
print(preco(lemonade_df, "Beach"))
```

```python
#9 - escreva uma função que receba o dataset como parâmetro e retorne o dataset com uma
#coluna a mais que mostre o rank de temperatura. O dataset retornado precisa estar ordenado
#de acordo com rank

def tempRank(ds):
    ds["rank"] = ds["Temperature"].rank(ascending=0)
    return ds.sort_values(by=['rank'])
```

```python
#14 - usando regressão linear, prediga vendas a partir de:
#- temperature
#- leafltes
#- price
#- retorne o gráfico com a reta ajustada aos dados para cada variável

import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from matplotlib import style
from statistics import mean

def predict_sell(df, feature_col, meta_col):
    X = df[feature_col].values
    y = df[meta_col].values
    split_test_size = 0.30 # 30% de treinamento
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_test_size, random_state=42)

    # Create linear regression object
    lr_model = linear_model.LinearRegression()

    # Train the model using the training sets
    lr_model.fit(X_train, y_train.ravel()) # flatten arrays

    # Make predictions using the testing set
    y_pred = lr_model.predict(X_test)
    
    # Print Results
    print_results_predict_sell(X_test, y_test, feature_col, y_pred)

def print_results_predict_sell(X_test, y_test, feature_col, y_pred):
    dfResult = pd.DataFrame(data=X_test)
    dfResult.columns = [feature_col]
    dfResult['Resultado'] = y_pred
    print(dfResult)
    
    # Graph line adjusted
    xs = X_test[:, 0]
    ys = y_test[:, 0]
    m, b = best_fit_slope_and_intercept(xs,ys)
    regression_line = [(m*x)+b for x in xs]
    
    #Plot de Graph
    style.use('ggplot')
    plt.title("Lucro por tipo de fruta")
    plt.xlabel(feature_col)
    plt.ylabel(meta_col)
    plt.scatter(xs,ys,color='#003F72')
    plt.plot(xs, regression_line)
    plt.show()
    print('\n')
    
def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b    

# By Temperature
feature_col = ['Temperature']
meta_col = ['Sales']
predict_sell(lemonade_df, feature_col, meta_col)

# By Leaflets
feature_col = ['Leaflets']
meta_col = ['Sales']
predict_sell(lemonade_df, feature_col, meta_col)

# By Price
feature_col = ['Price']
meta_col = ['Sales']
predict_sell(lemonade_df, feature_col, meta_col)
```

```python
#16) Usando a Regressão Linear implementada em sala como modelo, generalize o algoritmo 
# para que ele possa ser usado para regressão multivariada


class LinearRegression():
    
    def cost_function(self,x,y,theta):
        m = len(y) #número de amostras
        J = 0 # custo
        
        ypred = pd.DataFrame(np.matmul(x, np.transpose(theta)))
        
        for w in range(m):  
                J = J + np.power(( ypred.iloc[w].values - y.iloc[w].values),2)
                           
        return J*(1/(2*m))
    
    def gradient_descent(self,x,y,theta,alpha,num_iter):
        temp0 = 0
        iterations = num_iter
        theta_df = pd.DataFrame(theta)
        theta_df_temp = theta_df
        c = len(x.columns)
        m = len(y)
        J_history = pd.DataFrame(np.zeros(num_iter))
        
        y = pd.DataFrame(y)

        for k in range(iterations):        
                for i in range(m): 
                    ypred = pd.DataFrame(np.matmul(x, np.transpose(theta)))
                    for j in range(c):                       
                        if j == 0:
                            #Coeficiente linear        
                            theta_df_temp.iloc[j] = theta_df_temp.iloc[j] - alpha * (1/m * (ypred.iloc[i].values - y.iloc[i].values))
                            theta[j] = theta_df_temp.iloc[j] 
                            
                        else:
                            #Coeficiente Angular                          
                            theta_df_temp.iloc[j] = theta_df_temp.iloc[j] - alpha * (1/m * (ypred.iloc[i].values - y.iloc[i].values))*x.loc[i][j]
                            theta[j] = theta_df_temp.iloc[j]
                           
                            
                    J_history.iloc[k] = self.cost_function(x,y,theta)        
        return theta_df, J_history


# Exemplo de aplicação no Limonade    
    
feature_col = ['Temperature', 'Leaflets', 'Price']
meta_col = ['Sales']

x_lr = lemonade_df[feature_col]
ones = np.ones(len(x_lr))
x_lr.insert(0,len(x_lr.columns) + 1,ones)
y = lemonade_df[meta_col]
theta = np.zeros(len(x_lr.columns))

res_theta = pd.DataFrame(theta)
iterations = 20
custo =  pd.DataFrame(np.zeros(iterations))

alpha = 0.001

lr = LinearRegression()
res_theta , custo = lr.gradient_descent(x_lr,y,theta,alpha, iterations)

print("Theta: ", res_theta)
print("custo", custo)

```
